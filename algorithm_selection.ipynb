```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.svm import SVC
import pickle
```

```python
# Se carga los datos preprocesados
with open('/content/drive/MyDrive/Colab Notebooks/caracteristicas para datos medicos/preprocessedData.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
# Se separan las caracteristicas de las etiquetas
X = [d[:-1] for d in data]
y = [d[-1] for d in data]
```

```python
# Se dividen los datos entre entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
# Se prueba SVM
testCases = (0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 100 )
svmResults = []
for c in testCases:
  svm_model_linear = SVC(kernel = 'linear', C = c).fit(X_train, y_train)
  svm_predictions = svm_model_linear.predict(X_test)
  
  # model accuracy for X_test 
  accuracy = svm_model_linear.score(X_test, y_test)
  
  # creating a classification report
  r = classification_report(y_test, svm_predictions,output_dict=True)
  svmResults.append(r['accuracy'] )
print('--------------------------------')
print(max(svmResults))
```

```python
#Se prueba con KNN
from sklearn.neighbors import KNeighborsClassifier
testCases = list(range(1,300))
KnnResults = []
for n in testCases:
  knn = KNeighborsClassifier(n_neighbors = n).fit(X_train, y_train)
  # accuracy on X_test
  accuracy = knn.score(X_test, y_test) 
  # creating a classification report
  knn_predictions = knn.predict(X_test)
  r = classification_report(y_test, knn_predictions,output_dict=True)
  KnnResults.append(r['accuracy'] )
print('--------------------------------')
print(max(KnnResults))
```

```python
# training a Naive Bayes classifier
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB().fit(X_train, y_train)
gnb_predictions = gnb.predict(X_test)
# creating a classification report
cm = classification_report(y_test, gnb_predictions)
print(cm)
```

```python
#Se prueba con LogisticRegression
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver="lbfgs",max_iter=1000000)
clf.fit(X_train, y_train)
logistic_Pre=clf.predict(X_test)
# creating a classification report
cm = classification_report(y_test, logistic_Pre)
print(cm)
```
